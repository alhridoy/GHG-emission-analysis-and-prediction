```python
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
def reg_coef(x, y, label=None, color=None, **kwargs):
    ax = plt.gca()
    r, p = pearsonr(x, y)
    ax.annotate('r = {:.2f}'.format(r), xy=(0.5, 0.5), xycoords='axes fraction', ha='center',fontsize=20)
    
    # Add asterisks based on p-value
    if p < 0.001:
        ax.annotate('***', xy=(0.78, 0.53), xycoords='axes fraction', ha='center', fontsize=12)
    elif p < 0.01:
        ax.annotate('**', xy=(0.78, 0.53), xycoords='axes fraction', ha='center', fontsize=12)
    elif p < 0.05:
        ax.annotate('*', xy=(0.78, 0.53), xycoords='axes fraction', ha='center', fontsize=12)

    ax.set_axis_off()

# Assuming df is your DataFrame
g = sns.PairGrid(df)
g.map_diag(sns.histplot)  # Use histplot instead of distplot
g.map_lower(sns.regplot)
g.map_upper(reg_coef)
g.fig.suptitle('Pairwise Relationships',fontsize=35)  # Adjust the y parameter for vertical positioning
plt.savefig("C://Users//bijoy//Dropbox//PC//Desktop//Mitra//GHG_ML//6_corr.png", dpi = 300)
```
